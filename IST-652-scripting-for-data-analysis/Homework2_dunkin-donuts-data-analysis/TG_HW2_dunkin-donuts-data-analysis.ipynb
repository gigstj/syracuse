```python
#import libraries # use as needed

import json
import pandas as pd
import requests
import numpy as np
```

```python
#read the data from the url and create a dictionary object from there create a pd dataFrame.

url = 'https://www.dunkindonuts.com/bin/servlet/dsl'
response = requests.post(url, allow_redirects=False, data={
    'service': "DSL",
    'origin': '37.7749, -122.4194',
    'radius': "100000",
    'maxMatches': "100000"
})

content = json.loads(response.content)
newJson = {"data": []}

for entry in content["data"]["storeAttributes"]:
    entry["geoJson"] = json.loads(entry["geoJson"])
    newJson["data"].append(entry)

with open('dunkinDonuts.json', 'w') as jsonFile:
    json.dump(newJson, jsonFile)
```

```python
#view dict object
newJson
```

```python
#taking the dictionary object and create a pd dataframe
dunkindonuts_data=[]

for i in newJson.items(): #tuples for ease of parsing
    dunkindonuts_data.append([i[0],i[1]])
    
DF1=pd.DataFrame(dunkindonuts_data,columns=['dunkindonuts',' '])

#flatten lists to a flat table for easier read
DF1=DF1.explode(' ')

DF1.to_json(orient="records")

dunkinDonuts=pd.json_normalize(json.loads(DF1.to_json(orient="records")))
```

```python
#Question 1: Which state has the most dunkin donuts locations that offers mobile service?

Q1df = dunkinDonuts_table[[' .mobile', ' .state']]
Q1df
```

```python
# answer to question 1
# counts each time where Mobile = 'Y' for each state and puts into a dictionary

# start off by creating an empty dictionary
# this will be appended to after each loop iteration
statemobile = {}

# iterate through each state in a unique list of states
for tempstate1 in list(np.unique(Q1df[' .state'])):
    
    # start off by setting the mobile count as 0
    mobilecount = 0
    
    # iterate through each row of data
    # pack together the two elements with a hyphen
    # they will be unpacked again right after
    for packedtext in Q1df[' .state'] + '-' + Q1df[' .mobile']:
        
        # unpack the two elements by splitting on the hyphen
        # this creates a 2 element list
        # element 1 is the element left of the hyphen (state)
        # element 2 is the element right of the hyphen (mobile)
        unpackedtext = packedtext.split('-')
        
        # check if the state is equal to the state in the current iteration
        # check if the mobile is equal to Y
        if unpackedtext[0] == tempstate1 and unpackedtext[1] == 'Y':
            
            # if above conditions are met, add count + 1
            mobilecount += 1
    
    # add the state as the key and the count of mobiles as the value to the main dictionary
    statemobile[tempstate1] = mobilecount

# print out the resulting dictionary
print(statemobile)

# NY has the most with a count of 1340
# MA has the second most with a count of 1053
# NJ has the third most with 832
# FL has the fourth most with 822
# IL has the fifth most with 668
```

```python
# Question 2: Investigate the state and city with the highest number of sites offering delivery through GrubHub, Door Dash, and Uber Eats
# what is the average number of delivery options for those sites

Q2df = dunkinDonuts_table[[' .state', ' .city', ' .gh_active', ' .dd_active',' .ue_active']]
Q2df.head()
```

```python
# Column Reflecting Count of City Sites
Q2df['City_Sites'] = Q2df[' .city'].map(Q2df[' .city'].value_counts())
Q2df.head()
```

```python
# Column Reflecting Count of Delivery Methods per Site
Q2df['Delivery_Methods'] = (Q2df[[' .gh_active',' .dd_active',' .ue_active']] == 'Y').sum(axis=1)
Q2df.head()
```

```python
# Group by City - Count of Sites - Sum of Delivery Methods
Q2df2 = Q2df.groupby([' .city',' .state']).agg({'City_Sites':'count','Delivery_Methods':'sum'})
Q2df2.head()
```

```python
# Calculate the average number of delivery methods per site in each city
Q2df2['avg_delivery_methods'] = Q2df2['Delivery_Methods']/Q2df2['City_Sites']
Q2df2.head()
```

```python
# Sort data frame to identify city with the largest number of sites
Q2df2 = Q2df2.sort_values(by=['City_Sites'], ascending=False)

# Pull the top 10 cities
Top10Cities = Q2df2.head(10)
Top10Cities
```

```python
# Sort top 10 cities by Average Delivery Methods
Top10Cities.sort_values(by=['avg_delivery_methods'], ascending=False)
```
